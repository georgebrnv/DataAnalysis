```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column_names = ["DATE", "TAG", "POSTS"]
```

```python
df = pd.read_csv('QueryResults.csv', names=column_names, header=0)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.count()
```

```python
df.groupby("TAG").sum()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df.head()
```

```python
reshaped_df = df.pivot(index='DATE', columns='TAG', values='POSTS')
```

```python
reshaped_df
```

```python
reshaped_df.shape
```

```python
reshaped_df.columns
```

```python
reshaped_df.count()
```

```python
reshaped_df.fillna(0, inplace=True)
```

```python
reshaped_df.head()
```

```python
reshaped_df.isna().values.any()
```

```python
# The window is a number of observations that are averaged
roll_df = reshaped_df.rolling(window=7).mean()

plt.figure(figsize=(14, 7))
plt.xticks(fontsize=13)
plt.xlabel('Date', fontsize=13, color='red')
plt.yticks(fontsize=13)
plt.ylabel('Number of posts', fontsize=13, color='red')
plt.ylim(0, 37000)

# Plot all languages
for lang in roll_df.columns:
  plt.plot(roll_df.index, roll_df[lang], linewidth=2, label=roll_df[lang].name)

plt.legend(fontsize=10)
```
